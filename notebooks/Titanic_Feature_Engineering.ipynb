```python
import pandas as pd
df = pd.read_csv('../data/train.csv')
df.head()
```

```python
# Handle missing Age [cite: 33]
df['Age'] = df['Age'].fillna(df['Age'].median())

# Handle missing Embarked [cite: 33]
df['Embarked'] = df['Embarked'].fillna(df['Embarked'].mode()[0])

# Create indicator for Cabin [cite: 34]
df['Has_Cabin'] = df['Cabin'].apply(lambda x: 0 if pd.isna(x) else 1)

# Fix consistency and remove duplicates [cite: 40, 42]
df['Sex'] = df['Sex'].str.lower()
df = df.drop_duplicates()

# Save cleaned version [cite: 44]
df.to_csv('../data/train_cleaned.csv', index=False)
print("Cleaning Complete!")
```

```python
# Create FamilySize and IsAlone [cite: 51, 52]
df['FamilySize'] = df['SibSp'] + df['Parch'] + 1
df['IsAlone'] = df['FamilySize'].apply(lambda x: 1 if x == 1 else 0)

# Extract Title from Name [cite: 53]
df['Title'] = df['Name'].str.extract(' ([A-Za-z]+)\.', expand=False)

print(df[['Name', 'Title', 'FamilySize']].head())
```
